# Data Cleaning

```python
# Changing the event date data type to datetime
df["event_date"] = pd.to_datetime(df["event_date"], format="%Y-%m-%d")
```

```python
df["year"] = df["event_date"].dt.year
```

```python
df.head()
```

```python
# Summing null values in columns
null_columns=df.columns[df.isnull().any()]
for i in null_columns:
    print(i,df[i].isnull().sum())
```

```python
# Dropping colums with null values and irrelevant
df_cleaned = df.drop(columns=['admin3', 'timestamp','tags','event_id_cnty','time_precision',
                              'geo_precision','notes','source','source_scale','country','region','assoc_actor_1',
                              'assoc_actor_2','inter2','actor2','interaction','iso', "civilian_targeting"])
```

```python
# Checking for column unique values
for col in df_cleaned.columns:
    print(col, df_cleaned[col].nunique())
```

```python
df_cleaned[["admin1", "admin2"]]
```

```python
# Drop admin1 and admin2 with missing values
df_cleaned.dropna(subset=['admin1', 'admin2'], inplace=True)

# Check for duplicates and remove if found
df_cleaned.drop_duplicates(inplace=True)
```

```python
# save cleaned_dataframe to a new csv file
df_cleaned.to_csv('Nigeria_1997-2024_Sep20_cleaned.csv', index=False)
```
